import pandas as pd

from textblob import TextBlob

# Load the spaCy model for NER
nlp = spacy.load('en_core_web_sm')

# Load the filtered dataset
file_path = 'C:/Users/User/Desktop/Deakin/Year 3/T2/SIT374 Team Project A/Smishing/Merged_Dataset2.csv'
try:
    # Read the dataset from the specified path
    data = pd.read_csv(file_path, encoding='latin1')

    # Handle missing values in 'TEXT' column by filling with empty strings
    data['TEXT'] = data['TEXT'].fillna('')

    # Feature 1: Message Length
    data['message_length'] = data['TEXT'].apply(len)

    # Feature 2: Named Entity Recognition (NER)
    def count_named_entities(text):
        doc = nlp(text)
        # Count only certain entities
        return len([ent for ent in doc.ents if ent.label_ in ['PERSON', 'ORG', 'GPE']])

    # Apply NER to extract the number of named entities
    data['num_named_entities'] = data['TEXT'].apply(count_named_entities)

    # Feature 3: Sentiment Scores
    def get_sentiment_score(text):
        blob = TextBlob(text)
        return blob.sentiment.polarity  # Polarity score ranges from -1 (negative) to +1 (positive)

    # Calculate sentiment polarity for each message
    data['sentiment_score'] = data['TEXT'].apply(get_sentiment_score)

    # Save the updated dataset with new features

    print(f"Updated dataset saved as '{output_file_path}'.")

except FileNotFoundError as e:

Data Processing, Anomaly Detection, and Model Evaluation
Overview

In this phase of the project, I aimed to preprocess the dataset, detect anomalies, perform clustering, and build a supervised learning model for smishing detection. The results were evaluated using various metrics, and model explainability was enhanced with SHAP plots.
What I Did
1. Data Loading and Preprocessing

    Dataset Loading:
        I loaded the dataset from 'filtered_dataset_with_features.csv' and ensured it was in the correct format for processing.

    Handling Missing Values:
        Any missing values in the 'TEXT' column were replaced with empty strings to avoid issues during text cleaning.

    Text Cleaning:
        The text in the dataset was cleaned by converting it to lowercase and removing punctuation and special characters.

    Validation:
        I verified that no missing values remained in the 'TEXT_cleaned' column, ensuring the data was ready for further processing.

2. Feature Engineering and Dimensionality Reduction

    TF-IDF Vectorization:
        I applied TF-IDF (Term Frequency-Inverse Document Frequency) vectorization to transform the cleaned text data into numerical format. This allowed me to capture important terms while reducing the influence of common words.

    Dimensionality Reduction with Truncated SVD:
        I used Truncated Singular Value Decomposition (SVD) to reduce the dimensionality of the data to 100 components, making the data more manageable while retaining important information.

3. Anomaly Detection and Clustering

    Isolation Forest for Anomaly Detection:
        I used an Isolation Forest to identify potential anomalies in the dataset. This method helps detect outliers that could represent unusual or irrelevant messages.

    DBSCAN for Clustering:
        I applied DBSCAN (Density-Based Spatial Clustering of Applications with Noise) to group similar messages into clusters. This algorithm groups data points based on density, helping identify natural clusters of messages.

    Filtering Non-Useful Messages:
        Messages flagged as anomalies or belonging to trivial clusters (such as noise) were filtered out, leaving only useful and meaningful messages for model training.

4. Supervised Learning and Hyperparameter Tuning

    Train-Test Split:
        I split the data into training and test sets using an 80-20 ratio to ensure proper model evaluation.

    Random Forest Model with Hyperparameter Tuning:
        I built a Random Forest Classifier and used GridSearchCV to perform hyperparameter tuning. The parameters tuned included the number of estimators and the maximum depth of the trees.

5. Model Evaluation

    Classification Report:
        After training the model, I evaluated its performance on the test set using various metrics including precision, recall, F1-score, and support for each class. These metrics provided insights into how well the model performed across different categories (ham, spam, smishing).

6. Model Explainability with SHAP

    SHAP Summary Plot:
        I used SHAP (SHapley Additive exPlanations) to explain the model's predictions. SHAP provides insights into how each feature contributed to the model’s decision-making process.

    Visualization:
        A SHAP summary plot was generated to visualize the interaction and importance of features. This plot shows the most influential terms from the TF-IDF vectorizer that contributed to the model’s predictions.

7. Saving the Filtered Dataset

    Filtered Dataset:
        After processing, the filtered dataset (without anomalies and trivial clusters) was saved as 'filtered_dataset_advanced.csv' for future use.

Results
Classification Metrics:

    The classification report provides a breakdown of the model’s performance across different classes (Ham, Spam, Smishing). The precision, recall, and F1-score for each class are shown, helping assess how well the model distinguishes between these categories.

SHAP Summary Plot:

    The SHAP summary plot (shown in the image) highlights key terms like "cash," "1000," and "12hrs," which are significant in determining whether a message is ham, spam, or smishing.
    Each dot represents a prediction, with the colors representing the feature value (red for high, blue for low) and the horizontal axis representing the SHAP value (feature's contribution to the prediction).

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import DBSCAN
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import shap
import matplotlib.pyplot as plt

# Load and preprocess data
data = pd.read_csv('filtered_dataset_with_features.csv', encoding='latin1')

# Fill NaN values in 'TEXT' column with empty strings
data['TEXT'] = data['TEXT'].fillna('')

# Clean the text data
data['TEXT_cleaned'] = data['TEXT'].str.lower().str.replace(r'[^\w\s]', '', regex=True)

# Verify that there are no NaN values in 'TEXT_cleaned'
print("Number of NaN values in 'TEXT_cleaned':", data['TEXT_cleaned'].isnull().sum())

# Proceed with TF-IDF Vectorization and Dimensionality Reduction
tfidf = TfidfVectorizer(max_features=1000, ngram_range=(1, 3))
X_tfidf = tfidf.fit_transform(data['TEXT_cleaned'])
svd = TruncatedSVD(n_components=100)
X_reduced = svd.fit_transform(X_tfidf)


# Anomaly Detection and Clustering
iso_forest = IsolationForest(contamination=0.1, random_state=42)
anomalies = iso_forest.fit_predict(X_reduced)
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(X_reduced)

# Add anomaly and cluster information to the dataset
data['anomaly'] = anomalies
data['cluster'] = clusters

# Filter out non-useful messages (e.g., anomalies or trivial clusters)
filtered_data = data[(data['anomaly'] == 1) & (data['cluster'] != -1)]

# Supervised Learning: Model Training with Hyperparameter Tuning
X_train, X_test, y_train, y_test = train_test_split(X_reduced, data['LABEL'], test_size=0.2, random_state=42)
rf_model = RandomForestClassifier(random_state=42)
param_grid = {'n_estimators': [100, 200], 'max_depth': [10, 20, None]}
grid_search = GridSearchCV(rf_model, param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Model Evaluation and Explainability
y_pred = grid_search.best_estimator_.predict(X_test)
print(classification_report(y_test, y_pred))

# SHAP Summary Plot for Explainability
explainer = shap.TreeExplainer(grid_search.best_estimator_)
shap_values = explainer.shap_values(X_test)
plt.figure()
shap.summary_plot(shap_values, X_test, feature_names=tfidf.get_feature_names_out())
plt.savefig('shap_summary_plot.png', bbox_inches='tight')
print("SHAP summary plot saved as 'shap_summary_plot.png'.")

# Save the final filtered dataset
filtered_data.to_csv('filtered_dataset_advanced.csv', index=False)
print("Advanced filtered dataset saved as 'filtered_dataset_advanced.csv'.")


```
